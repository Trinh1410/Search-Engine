```python
from pyspark.sql import SparkSession

# Tạo Spark session
spark = SparkSession.builder \
    .appName("HiveExample") \
    .config("spark.sql.catalogImplementation", "hive") \
    .enableHiveSupport() \
    .getOrCreate()

# Tạo database mới (nếu chưa có)
spark.sql("CREATE DATABASE IF NOT EXISTS my_database")

# Chọn database mới
spark.sql("USE my_database")

# Tạo bảng khách hàng
spark.sql("""
    CREATE TABLE IF NOT EXISTS customers (
        customer_id INT,
        customer_name STRING,
        contact_name STRING,
        country STRING
    )
    ROW FORMAT DELIMITED
    FIELDS TERMINATED BY ','
    STORED AS TEXTFILE
""")

# Tạo bảng đơn hàng
spark.sql("""
    CREATE TABLE IF NOT EXISTS orders (
        order_id INT,
        order_date STRING,
        customer_id INT,
        amount DOUBLE
    )
    ROW FORMAT DELIMITED
    FIELDS TERMINATED BY ','
    STORED AS TEXTFILE
""")

# Tạo bảng sản phẩm
spark.sql("""
    CREATE TABLE IF NOT EXISTS products (
        product_id INT,
        product_name STRING,
        supplier_id INT,
        category STRING,
        price DOUBLE
    )
    ROW FORMAT DELIMITED
    FIELDS TERMINATED BY ','
    STORED AS TEXTFILE
""")

# Chèn dữ liệu vào bảng khách hàng
spark.sql("""
    INSERT INTO TABLE customers VALUES
    (1, 'Alfreds Futterkiste', 'Maria Anders', 'Germany'),
    (2, 'Centro comercial Moctezuma', 'Francisco Chang', 'Mexico'),
    (3, 'Ernst Handel', 'Roland Mendel', 'Austria')
""")

# Chèn dữ liệu vào bảng đơn hàng
spark.sql("""
    INSERT INTO TABLE orders VALUES
    (101, '2024-01-01', 1, 150.00),
    (102, '2024-02-15', 2, 200.00),
    (103, '2024-03-10', 3, 250.00)
""")

# Chèn dữ liệu vào bảng sản phẩm
spark.sql("""
    INSERT INTO TABLE products VALUES
    (1, 'Chai', 1, 'Beverages', 18.00),
    (2, 'Chang', 1, 'Beverages', 19.00),
    (3, 'Aniseed Syrup', 2, 'Condiments', 10.00)
""")

# Kiểm tra dữ liệu trong bảng khách hàng
spark.sql("SELECT * FROM customers").show()

# Kiểm tra dữ liệu trong bảng đơn hàng
spark.sql("SELECT * FROM orders").show()

# Kiểm tra dữ liệu trong bảng sản phẩm
spark.sql("SELECT * FROM products").show()

# Kết thúc Spark session
spark.stop()
```

```python
from pyspark.sql import SparkSession

# Tạo Spark session
spark = SparkSession.builder \
    .appName("HiveExample") \
    .config("spark.sql.catalogImplementation", "hive") \
    .enableHiveSupport() \
    .getOrCreate()

# Tạo cơ sở dữ liệu
spark.sql("CREATE DATABASE IF NOT EXISTS student")

# Chọn cơ sở dữ liệu
spark.sql("USE student")

# Tạo bảng teacher
spark.sql("""
    CREATE TABLE IF NOT EXISTS teacher (
        id INT,
        name STRING,
        subject STRING
    )
    ROW FORMAT DELIMITED
    FIELDS TERMINATED BY ','
    STORED AS TEXTFILE
""")

# Tạo bảng student
spark.sql("""
    CREATE TABLE IF NOT EXISTS student (
        id INT,
        name STRING,
        age INT,
        teacher_id INT
    )
    ROW FORMAT DELIMITED
    FIELDS TERMINATED BY ','
    STORED AS TEXTFILE
""")

# Tạo bảng course
spark.sql("""
    CREATE TABLE IF NOT EXISTS course (
        id INT,
        name STRING,
        teacher_id INT
    )
    ROW FORMAT DELIMITED
    FIELDS TERMINATED BY ','
    STORED AS TEXTFILE
""")

# Chèn dữ liệu vào bảng teacher
spark.sql("""
    INSERT INTO TABLE teacher VALUES
    (1, 'Alice', 'Mathematics'),
    (2, 'Bob', 'Physics'),
    (3, 'Charlie', 'Chemistry')
""")

# Chèn dữ liệu vào bảng student
spark.sql("""
    INSERT INTO TABLE student VALUES
    (1, 'David', 20, 1),
    (2, 'Eve', 22, 2),
    (3, 'Frank', 21, 3)
""")

# Chèn dữ liệu vào bảng course
spark.sql("""
    INSERT INTO TABLE course VALUES
    (1, 'Calculus', 1),
    (2, 'Quantum Mechanics', 2),
    (3, 'Organic Chemistry', 3)
""")

# Kiểm tra dữ liệu trong bảng teacher
spark.sql("SELECT * FROM teacher").show()

# Kiểm tra dữ liệu trong bảng student
spark.sql("SELECT * FROM student").show()

# Kiểm tra dữ liệu trong bảng course
spark.sql("SELECT * FROM course").show()

# Kết thúc Spark session
spark.stop()
```

```python
import happybase

# Kết nối đến HBase
connection = happybase.Connection('localhost')

# Tạo bảng
connection.create_table('student', {'info': dict()})

# Lấy bảng
table = connection.table('student')

# Chèn dữ liệu
table.put(b'1', {b'info:name': b'David', b'info:age': b'20'})
table.put(b'2', {b'info:name': b'Eve', b'info:age': b'22'})
table.put(b'3', {b'info:name': b'Frank', b'info:age': b'21'})

# Kiểm tra dữ liệu
for key, data in table.scan():
    print(key, data)
```
